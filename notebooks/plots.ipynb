```julia
# 读取数据
data_crisfield = load("crisfield_results.jld2")
disp_crisfield = -data_crisfield["disp"];
force_crisfield = -data_crisfield["force"];
gf_energy_crisfield = data_crisfield["gf_energy"];

data_gamma = load("gamma_results.jld2")
disp_gamma = -data_gamma["disp"];
force_gamma = -data_gamma["force"];
gf_energy_gamma = data_gamma["gf_energy"];

data_h1 = load("h1_results.jld2")
disp_h1 = -data_h1["disp"];
force_h1 = -data_h1["force"];
gf_energy_h1 = data_h1["gf_energy"];

data_l2 = load("l2_results.jld2")
disp_l2 = -data_l2["disp"];
force_l2 = -data_l2["force"];
gf_energy_l2 = data_l2["gf_energy"];

data_mem = load("mem_results.jld2")
disp_mem = -data_mem["disp"];
force_mem = -data_mem["force"];
gf_energy_mem = data_mem["gf_energy"];

data_sem = load("sem_results.jld2")
disp_sem = -data_sem["disp"];
force_sem = -data_sem["force"];
gf_energy_sem = data_sem["gf_energy"];

data_staggered = load("staggered_results.jld2")
disp_staggered = -data_staggered["disp"];
force_staggered = -data_staggered["force"];
gf_energy_staggered = data_staggered["gf_energy"];
```

```julia
# ==================== 第一份图：Disp vs. Force ====================
fig1 = Figure(size = (650, 450), fontsize = 22, fonts = (; regular = "Times New Roman"))
ax1 = Axis(fig1[1, 1],
    xlabel = L"$\bar{u}$ [mm]",
    ylabel = L"$F_{\text{reac}}$ [N]",
    limits = ((0, 0.8), (0, 350)),
    xgridvisible = true, ygridvisible = true,
    xgridstyle = :dash, ygridstyle = :dash
)
lines!(ax1, disp_crisfield, force_crisfield, color = :black, linestyle = :solid, linewidth = 1.2, label = "Crisfield")
lines!(ax1, disp_gamma, force_gamma, color = :black, linestyle = :dash, linewidth = 1.2, label = L"$\Gamma$")
scatterlines!(ax1, disp_h1, force_h1, color = :black, linestyle = :solid, marker = :circle, markersize = 4, linewidth = 1.2, label = L"$H_1$")
lines!(ax1, disp_l2, force_l2, color = :black, linestyle = :dashdot, linewidth = 1.2, label = L"$L_2$")
lines!(ax1, disp_mem, force_mem, color = :red, linestyle = :solid, linewidth = 1.2, label = "M-EM")
lines!(ax1, disp_staggered, force_staggered, color = :red, linestyle = :dash, linewidth = 1.2, label = "Staggered")
lines!(ax1, disp_sem, force_sem, color = :red, linestyle = :dot, linewidth = 1.2, label = "S-EM")

axislegend(ax1; 
    position = :lt,             # 位置：左上角
    labelsize = 16,             # 标签字体大小调为 14 (比坐标轴字体 18 略小，更显精致)
    patchsize = (40, 10),       # 【核心】把图例线条拉长到 40，这样虚线/点划线特征非常明显
    rowgap = 2,                 # 紧凑排版：减小行间距
    patchlabelgap = 6,          # 减小线条与文字的间距
    padding = (8, 8, 6, 6),     # 紧凑排版：减小图例边框内边距
    backgroundcolor = (:white, 0.85)     # 背景微透明，防止完全遮挡后面的网格线
)

# 保存第一张图
save("disp_vs_force.pdf", fig1)


# ==================== 第二份图：Disp vs. gf_energy ====================
fig2 = Figure(size = (650, 450), fontsize = 22, fonts = (; regular = "Times New Roman"))
ax2 = Axis(fig2[1, 1],
    xlabel = L"$\bar{u}$ [mm]",
    ylabel = L"$\mathcal{G}_f$ [Nmm]",
    xgridvisible = true, ygridvisible = true,
    limits = ((0, 0.8), (0, 82)),
    xgridstyle = :dash, ygridstyle = :dash
)
lines!(ax2, disp_crisfield, gf_energy_crisfield, color = :black, linestyle = :solid, linewidth = 1.2, label = "Crisfield")
lines!(ax2, disp_gamma, gf_energy_gamma, color = :black, linestyle = :dash, linewidth = 1.2, label = L"$\Gamma$")
scatterlines!(ax2, disp_h1, gf_energy_h1, color = :black, linestyle = :solid, marker = :circle, markersize = 4, linewidth = 1.2, label = L"$H_1$")
lines!(ax2, disp_l2, gf_energy_l2, color = :black, linestyle = :dashdot, linewidth = 1.2, label = L"$L_2$")
lines!(ax2, disp_mem, gf_energy_mem, color = :red, linestyle = :solid, linewidth = 1.2, label = "M-EM")
lines!(ax2, disp_staggered, gf_energy_staggered, color = :red, linestyle = :dash, linewidth = 1.2, label = "Staggered")
lines!(ax2, disp_sem, gf_energy_sem, color = :red, linestyle = :dot, linewidth = 1.2, label = "S-EM")

axislegend(ax2; 
    position = :lt, 
    labelsize = 16, 
    patchsize = (40, 10), 
    rowgap = 2, 
    patchlabelgap = 6, 
    padding = (8, 8, 6, 6), 
    backgroundcolor = (:white, 0.85)
)

# 保存第二张图
save("disp_vs_energy.pdf", fig2)
```
